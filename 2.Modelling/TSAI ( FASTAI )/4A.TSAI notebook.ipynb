```python
x_t =np.moveaxis(x_t,0,1)
print(x_t.shape)
```

```python
x_t=np.reshape(x_t,(x_t.shape[0],1,x_t.shape[1]))
print(x_t.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x_t,y,test_size=0.33)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
X_train=np.reshape(X_train,(X_train.shape[0],1,X_train.shape[1]))
#Y_train=np.reshape(Y_train,(Y_train.shape[0], 1,Y_train.shape[1]))
X_test = np.reshape(X_test,(X_test.shape[0],1,X_test.shape[1]))
#Y_test =np.reshape(Y_test,(Y_test.shape[0],1,Y_test.shape[1]))
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
for i in range(10):
    plt.figure()
    plt.plot(X_train[i])
    plt.title(Y_train[i])
```

```python
from tsai.all import *
clf =TSClassifier(X_train,Y_train,metrics=[f1score_binary,accuracy,Precision(average='micro'),Recall(average='micro')],batch_tfms=TSStandardize(), arch=InceptionTimePlus,arch_config=dict(fc_dropout=.5))
```

```python
f1score_binary = F1Score()
clf_adv =TSClassifier(X_train,Y_train,loss_func=DiceLoss(),metrics=[f1score_binary,accuracy,Precision(average='micro'),Recall(average='micro')], batch_tfms=TSStandardize(),arch=InceptionTimePlus,arch_config=dict(fc_dropout=.5))
```

```python
clf.summary()
```

```python
clf_adv.summary()
```

```python
lr = clf.lr_find()
```

```python
lr
```

```python
clf.fit_one_cycle(10,lr)
```

```python
clf.recorder.plot_metrics()
```

```python
_,_,preds = clf.get_X_preds(X_test)
```
