```python
#Sliding Window Segmentation
import itertools
def sliding_dist(benchmark_data, data):
  """
  data : input data should be data['i'] where i is column name
  return distance 
  """
  x = list(np.arange(0,len(data),3))
  y =[i+len(benchmark_data) for i in x]
  distance = []# Creating Distance Value by appending Dynamic Time Warping data
  for (i,j) in zip(x,y):
    if j<=len(data):
      distance.append(dissVV(benchmark_data,np.nan_to_num(data[i:j].interpolate())))
      fig,(ax1,ax2)= plt.subplots(2,1)
      plt.figure()
      ax1.plot(data[i:j].interpolate(),color='red',label=f"time window is from {i} to {j}")
      ax1.set_title(label=f'Distance is {dissVV(benchmark_data,np.nan_to_num(data[i:j].interpolate()))}')
      #ax1.set_title(label=f'Distance is {distance}')
      ax1.legend()
      ax2.plot(benchmark_data,label="Label Dataset")
      ax2.legend()
      
  return(distance)
      #fig.savefig(f'/content/plots/timestamps{i}&{j}.png')
```

```python
def plotting_subset(x,y):
  plt.plot(range(50),sample[x:y].interpolate())

def polynomial_calc(x,y):
  curve =np.polyfit(range(50),np.nan_to_num(sample[x:y].interpolate()),4)
  poly=np.poly1d(curve)
  print(poly)
  plt.figure(figsize=(20,10))
  plt.plot(np.array(sample[x:y].interpolate()),color='red',label='Original data');
  plt.plot(poly(range(50)),color='blue',label='Polynomial Fit');
  plt.legend()
  #plt.title(label=poly)
  return poly
```

```python
#Calculating polynomial coefficients and creating a list of it
x = list(np.arange(0,len(sample),3))
y =[i+len(benchmark) for i in x]
polyN = []
for (i,j) in zip(x,y):
  if j<=len(sample):
    polyN.append(polynomial_calc(i,j))
```
